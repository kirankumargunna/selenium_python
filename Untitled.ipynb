```python
import requests
import json

# Replace with your own API key
api_key = 'da97c210'
url = f"http://www.omdbapi.com/?apikey={api_key}&s=movie&plot=short"

# Sending a GET request to fetch data
response = requests.get(url)

# Checking if the response was successful
if response.status_code == 200:
    data = response.json()
    
    # Check if there are results
    if data['Response'] == 'True':
        # List of movie titles and their IMDb IDs
        movies = data['Search']
        
        for movie in movies:
            print(f"Title: {movie['Title']}, Year: {movie['Year']}, IMDb ID: {movie['imdbID']}")
    else:
        print("No results found.")
else:
    print(f"Error fetching data: {response.status_code}")
```

```python
import requests

# Replace with your actual API key
api_key = 'da97c210'

page = 1
while True:
    # Construct the URL with the correct API key and pagination
    url = f"http://www.omdbapi.com/?apikey={api_key}&s=movie&page={page}&plot=short"
    
    # Send the GET request
    response = requests.get(url)
    
    # Parse the response
    data = response.json()
    
    # Check if the response is successful and contains movies
    if data['Response'] == 'True':
        movies = data['Search']
        
        # Print out the movie details
        for movie in movies:
            print(f"Title: {movie['Title']}, Year: {movie['Year']}, IMDb ID: {movie['imdbID']}")
        
        # Check if there are more pages
        if len(movies) == 0:
            print("No more movies found.")
            break  # Exit the loop if there are no more movies
        
        page += 1  # Increment to the next page
    
    else:
        # If no response or an error occurs
        print("No more data or an error occurred:", data.get('Error', 'Unknown error'))
        break  # Exit the loop
```
