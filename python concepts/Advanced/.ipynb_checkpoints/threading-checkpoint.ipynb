```python
import time

def calculate_square(numbers):
    print("Calculate square")
    for n in numbers:
        print("Square:", n*n)
        
def calculate_cube(numbers):
    print("Calculate cube")
    for n in numbers:
        print("Cube:", n**3)

lst = list(range(1, 10000, 10))
t = time.time()

# Sequential execution
calculate_square(lst)
calculate_cube(lst)

print("Time taken:", time.time() - t)
```

```python
import time
import threading

def calculate_square(numbers):
    print("Calculate square")
    for n in numbers:
        print("Square:", n*n)
        
def calculate_cube(numbers):
    print("Calculate cube")
    for n in numbers:
        print("Cube:", n**3)

lst = list(range(1, 10000, 10))
t = time.time()

# Create multiple threads
t1 = threading.Thread(target=calculate_square, args=(lst,))
t2 = threading.Thread(target=calculate_cube, args=(lst,))

# Start threads
t1.start()
t2.start()

# Wait for both threads to finish
t1.join()
t2.join()

print("Time taken:", time.time() - t)
```

```python
import time
import threading

#create a tastk 

def simulate_task(taskid):
    print(f"task{taskid} is started")
    time.sleep(1)
    print(f"task{taskid} is compleated")

# list of tasks
tasks=list(range(1,5))

#without threading
def withoutthreading():
    start_time=time.time()
    for task in tasks:
        simulate_task(task)
    print("without threading time is ", time.time()-start_time)


def with_threads():
    start_time=time.time()
    threads=[]
    for task in tasks:
        thread=threading.Thread(target=simulate_task,args=(task,))
        threads.append(thread)
        thread.start()

    for thread in threads:
        thread.join()
    print("Threaded execution time:", time.time() - start_time)


# Run without and with threaded executions
print("Running sequential execution...")
withoutthreading()

print("\nRunning threaded execution...")
with_threads()
```
