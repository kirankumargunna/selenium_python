```python
class X:
    def __init__(self):
        self.a = 1
        self.b = 2
        self.lock = threading.RLock()

    def changeA(self):
        with self.lock:
            self.a = self.a + 1
            print(self.a)

    def changeB(self):
        with self.lock:
            self.b = self.b + 1
            print(self.b)

    def changeAandB(self):
        # you can use chanceA and changeB thread-safe!
        with self.lock:
            self.changeA() # a usual lock would block at here
            self.changeB()

x=X()

for i in range(5):
    thread=threading.Thread(target=x.changeAandB())
    thread.start()
    
```

```python
import threading
import time

# Lock definitions
lock1 = threading.Lock()
lock2 = threading.Lock()

# Thread 1: locks lock1 first, then lock2
def thread1():
    with lock1:
        print("Thread 1: Acquired lock1")
        time.sleep(1)  # Simulate some work
        print("Thread 1: Trying to acquire lock2")
        with lock2:
            print("Thread 1: Acquired lock2 and performing task")

# Thread 2: locks lock2 first, then lock1
def thread2():
    with lock2:
        print("Thread 2: Acquired lock2")
        time.sleep(1)  # Simulate some work
        print("Thread 2: Trying to acquire lock1")
        with lock1:
            print("Thread 2: Acquired lock1 and performing task")

# Create threads
t1 = threading.Thread(target=thread1)
t2 = threading.Thread(target=thread2)

# Start threads
t1.start()
t2.start()

# Wait for threads to finish
t1.join()
t2.join()

print("Both threads have finished.")


# this will create a dead lock  
```

```python
# lock in selenium 

import threading
import time
from selenium import webdriver

# Create a lock to control access to the WebDriver
lock = threading.Lock()

# Function to simulate a task performed by a thread using WebDriver
def thread_task(driver, thread_id):
    with lock:  # Ensures only one thread can interact with the WebDriver at a time
        print(f"Thread-{thread_id}: Starting task...")
        driver.get("https://www.example.com")
        time.sleep(2)  # Simulate some delay (e.g., waiting for page to load)
        print(f"Thread-{thread_id}: Finished task!")
        
# Function to launch a thread with a WebDriver instance
def create_driver_and_run_thread(thread_id):
    driver = webdriver.Chrome()
    # Create a new WebDriver instance (one per thread)
    thread_task(driver, thread_id)
    driver.quit()  # Quit the driver after use

# List to store thread references
threads = []

# Start multiple threads
for i in range(5):
    print(i)
    thread = threading.Thread(target=create_driver_and_run_thread, args=(i+1,))
    threads.append(thread)
    thread.start()
    

# Wait for all threads to complete
for t in threads:
    t.join()

print("All tasks completed.")
```
