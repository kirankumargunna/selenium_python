## Arithmetic Operators


```python
# Here are some basic arthematic operations 
x=int(input())
y=int(input())

print("Addition:", x + y)       
print("Subtraction:", x - y)    
print("Multiplication:", x * y) 
print("Division:", x / y)       
print("Floor Division:", x // y)
print("Modulus:", x % y)        
print("Exponentiation:", x ** y)
```

## comparision operator 
 comparision operators used to compare two values and the return type is boolean type 
 


```python
# Comparison Operators
x=int(input())
y=int(input())

print("Equal to:", x == y)       
print("Not equal to:", x != y)   
print("Greater than:", x > y)    
print("Less than:", x < y)       
print("Greater than or equal to:", x >= y)  
print("Less than or equal to:", x <= y)     
```

### comparision operators for strings 

```python
# Define some example strings
str1 = "hello"
str2 = "world"
str3 = "hello"
str4 = "apple"
str5 = "banana"
str6 = "Apple"
str7 = "abc"
str8 = "abcd"
str9 = "123"
str10 = "45"

# Equality and Inequality
print(f"Equality: {str1} == {str3} -> {str1 == str3}")  
print(f"Inequality: {str1} != {str2} -> {str1 != str2}")  

# Relational Operators
print(f"Less than: {str4} < {str5} -> {str4 < str5}")      
print(f"Greater than: {str5} > {str4} -> {str5 > str4}")  
print(f"Less than or equal: {str4} <= {str3} -> {str4 <= str3}")  
print(f"Greater than or equal: {str3} >= {str1} -> {str3 >= str1}")  

# Lexicographical Comparison
print(f"Comparing 'abc' < 'abd': {'abc' < 'abd'}")        
print(f"Comparing 'apple' > 'Apple': {'apple' > 'Apple'}") 
print(f"Comparing '123' < '45': {'123' < '45'}")          
```

## logical operators
Python logical operators (and, or, not) are special keywords that act as conjunctions in our code, connecting and evaluating multiple conditions within a single expression.


```python
# Ask user for their age and whether they have a license
age = int(input("Enter your age: "))
has_license = input("Do you have a driving license? (yes/no): ").lower() == 'yes'

if age >= 18 and has_license:
    print("You are eligible to drive.")
elif age >= 18 and not has_license:
    print("You are old enough to drive but need a license.")
else:
    print("You are not eligible to drive.")
```

## Assignment operators 
Used to assign values to variables



```python
# Assignment Operators
x = int(input())

# Add 5 to x
x += 5   # Same as x = x + 5
print("x after x+= 5:", x)

# Subtract 3 from x
x -= 3   # Same as x = x - 3
print("x after x-= 3:", x)

# Multiply x by 2
x *= 2   # Same as x = x * 2
print("x after x*= 2:", x)

# Divide x by 4
x /= 4   # Same as x = x / 4
print("x after x/= 4:", x)

# Raise x to the power of 2
x **= 2  # Same as x = x ** 2
print("x after x**= 2:", x)

# Floor divide x by 6
x //= 6  # Same as x = x // 6
print("x after x//= 6:", x)
```
