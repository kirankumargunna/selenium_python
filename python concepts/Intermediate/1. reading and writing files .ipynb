### Reading and writing files

Differnet modes of file handeling   

'r' – Read (default mode).  
'w' – Write (truncates the file if it exists).  
'a' – Append (writes data at the end of the file).  
'b' – Binary mode (for binary files like images).  
't' – Text mode (default mode for text files).  
'x' – Exclusive creation (fails if the file exists).  


__reading line by line__

```python
try:
    for line in file:
        print(line.strip())
    lines=file.readlines()
    print(lines)
finally:
    file.close()
```
