### Download the WebDriver for Your Browser
Selenium requires a WebDriver to control the browser. A WebDriver is a tool that interacts with the browser for your Selenium script. Depending on the browser you are using, you'll need to download the correct WebDriver  

download the latest stable version of chrome Webdriver. [ChromeDriver](https://googlechromelabs.github.io/chrome-for-testing/#stable)  

After downloading, extract the `chromedriver.exe` file  

### set up chrome driver path  

#### method 1: specify path in script

```python
from selenium import webdriver #import webdriver class from selenium package

from selenium.webdriver.chrome.service import Service    #import service class from module selenium.webdriver.chrome.service
# Specify the path to chromedriver
#service = webdriver.ChromeService(executable_path=r"C:\Users\Kiran Gunna\Downloads\chromedriver-win64\chromedriver.exe")

service = Service(executable_path=r"C:\Users\Kiran Gunna\Downloads\chromedriver-win64\chromedriver.exe")
driver=webdriver.Chrome(service=service)
# Example usage
driver.get('https://www.pythontutorial.net/')

# Close the browser
driver.quit()
```

### Method 2: Add to System PATH
Search for "Environment Variables" in the Start menu and open it.  
Click on "Environment Variables."  
In the "System variables" section, find and select the Path variable, then click "Edit."  
Click "New" and add the path to the folder where chromedriver.exe is located.  
Click OK to save changes
